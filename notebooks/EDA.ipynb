# Exploratory Data Analysis (EDA)

This notebook covers the exploratory data analysis for the weekly sales dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Dataset


```python
df = pd.read_csv('../data/Assessment-2-Associate-DS(in).csv')
df.head()
```

# Missing Values

```python
df.isnull().sum()
```

# Check Data Type and Basic Info

```python
df.info()
```

# Describe Numeric Values

```python
df.describe()
```

```python
df['weekend_date'] = pd.to_datetime(df['weekend_date'], dayfirst=True, errors='coerce')
weekly_sales = df.groupby('weekend_date')['quantity'].sum()
plt.figure(figsize=(12,6))
weekly_sales.plot()                             
plt.title('Weekly Sales Over Time')
plt.xlabel('Week Ending Date')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
plt.title('Weekly Sales Over Time')
plt.xlabel('Week Ending Date')
plt.ylabel('Quantity Sold')
plt.grid(True)
plt.show()     
```

# Check for outliers in Quantity


```python
sns.boxplot(x=df['quantity'])
plt.title('Boxplot of Weekly Quantity Sold')
plt.show()
```

```python
df['month'] = df['weekend_date'].dt.month
monthly_avg = df.groupby('month')['quantity'].mean()
plt.figure(figsize=(8,4))
monthly_avg.plot(kind='bar')
plt.title('Average Weekly Sales by Month')
plt.xlabel('Month')
plt.ylabel('Avg Quantity Sold')
plt.show()
   
```

# sales dist. by channels

```python
plt.figure(figsize=(10,4))
sns.boxplot(x='channel', y='quantity', data=df)
plt.title('Sales Distribution by Channel')
plt.show()
```

# sales distribution by brands


```python
plt.figure(figsize=(10,4))
sns.boxplot(x='brand', y='quantity', data=df)
plt.title('Sales Distribution by Brand')
plt.show()
```
